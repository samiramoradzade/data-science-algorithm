```python
#name har film ra dar listi b name BasketName rikhtim
BasketNames = []
for i in range (0 , 1000):
    BasketNames.append(Data.values[i][0])

#bazigaran har film ra b onvane ozvi az araye Basket dar nazar migirim
Basket = []
for i in range (0 , 1000):
    Basket.append(Data.values[i][1].split(','))
```

```python
#chon barkhi az esma b onvan string daraye space ezafe bodand ba dastor zir in space haye ezafe ra hazf kardim.
fixedBasket=[]
for i in Basket:
    fixedBasket.append([x.strip() for x in i])
Basket = fixedBasket
```

```python
#niyaz be esm bazigaran b tor jodagane dashtim k esm tekrari daron an nabashad banabar in listi khali tarif kardim va esmha ra done done b an ezafe kardim ba tavajoh b inke tekrari dakhel an gharar nagirad
Item = []
for i in range (0 , 1000):
    for j in range (len(Basket[i])):
        if Basket[i][j] not in Item:
            Item.append(Basket[i][j])

#baraye inke listi az bazigaran b hamrah tedad filmahayi k dar an bazi kardan ra dashte bashim ba dastor zir an hara ebteda shemordim va done done b dictionary khali k tarif kardim ezafe kardim            
SupDic = {}
for i in range (len(Item)):
    k = 0
    for j in range (0 , len(Basket)):
        if Item[i] in Basket[j]:
            k+=1
    SupDic[Item[i]] = k

print ("you can see actor names with the count of the film they had played:" , SupDic)
```

```python
#define the treshold = 2
#find the frequent items
FreqItems = {}
for Item in SupDic:
    if (SupDic[Item]>=2):
        FreqItems[Item] = SupDic[Item]
#hal mitavanim esm bazigarani k dar bish az 2 film az in 1000 film bartar bazi kardand ra bbinim  
print (FreqItems)
```
